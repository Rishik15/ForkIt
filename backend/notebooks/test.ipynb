```python
with open('../app/models/svd_model.pkl', 'rb') as f:
    svd = pkl.load(f)
```

```python
user_vector_reduced = svd.transform(user_df)
user_vector_reduced
```

```python
restaurantVectors = restaurant_vectors_df.iloc[:, 2:].values
restaurantVectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

cosine_similarities = cosine_similarity(user_vector_reduced, restaurantVectors)
top_5_indices = np.argsort(cosine_similarities[0])[::-1][:5]
```

```python
top_5_restaurants = restaurant_vectors_df.iloc[top_5_indices][['Place_ID', 'Name']]
```

```python
top_5_restaurants['Similarity'] = cosine_similarities[0][top_5_indices]
top_5_restaurants
```
