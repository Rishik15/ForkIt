```python
import pandas as pd
import re
import shutil
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt_tab')
shutil.rmtree('/root/nltk_data/tokenizers/punkt', ignore_errors=True)
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def process_sentences(text):
    """Advanced text preprocessing with lemmatization, POS tagging, stopword removal."""
    if pd.isna(text):
        return ""

    text = text.replace("n't", " not").replace("'m", " am") \
               .replace("'s", " is").replace("'re", " are") \
               .replace("'ll", " will").replace("'ve", " have") \
               .replace("'d", " would")

    temp_sent = []

    words = word_tokenize(text.lower())
    tags = nltk.pos_tag(words)

    for i, word in enumerate(words):
        if tags[i][1] in ('VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'):
            lemmatized = lemmatizer.lemmatize(word, 'v')
        else:
            lemmatized = lemmatizer.lemmatize(word)

        if lemmatized not in stop_words and lemmatized.isalpha():
            temp_sent.append(lemmatized)

    full_sentence = ' '.join(temp_sent)

    return full_sentence
```

```python
price_map = {
    1 : "cheap inexpensive low-price low-cost economical economic affordable",
    2 : "moderate fair mid-price reasonable average",
    3 : "expensive fancy lavish",
    4 : "luxurious high-end premium fine-dining"
}

df["Cuisine"] = df["Cuisine"].apply(lambda x: ', '.join(x) if isinstance(x, list) else str(x).replace("[","").replace("]","").replace("'",""))

df["Reviews"] = df["Reviews"].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x).replace("[","").replace("]","").replace("'",""))

df["Price_Level_Description"] = df["Price_Level"].map(price_map).fillna("")

df["Combined_Text"] = (
    df["Name"].fillna("") + " " +
    df["Editorial_Summary"].fillna("") + " " +
    df["Reviews"].fillna("") + " " +
    df["Cuisine"].fillna("") + " " +
    df["Price_Level_Description"]
)
```

```python
df['Combined_Text'][464]
```

```python
df.loc[:, "Processed_Text"] = df["Combined_Text"].apply(process_sentences)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(df["Processed_Text"])

print("TF-IDF vectorization complete. Shape:", tfidf_matrix.shape)
```

```python
import pickle

with open("/content/drive/MyDrive/ForkIt/tfidf_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

with open("/content/drive/MyDrive/ForkIt/tfidf_matrix.pkl", "rb") as f:
    tfidf_matrix = pickle.load(f)
```

```python
def process_user_query(query):
    """Preprocess and convert user query into a TF-IDF vector."""
    query_tfidf = vectorizer.transform([query])
    return query_tfidf
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_recommendations(user_query, top_n=5):
    """Finds the most similar restaurants based on a text query."""
    query_tfidf = process_user_query(user_query)

    similarity_scores = cosine_similarity(query_tfidf, tfidf_matrix)

    top_indices = similarity_scores.argsort()[0][-top_n:][::-1]

    return df.iloc[top_indices][["Name", "Cuisine", "Price_Level", "Editorial_Summary"]]
```

```python
user_query = "halal not pricy and the food tastes good"
recommendations = get_recommendations(user_query)
recommendations
```
